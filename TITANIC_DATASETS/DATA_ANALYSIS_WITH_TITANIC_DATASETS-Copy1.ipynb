# Data Explanation
<ul>
<li>survival - Survival (0 = No; 1 = Yes)
<li>class - Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)
<li>name - Name
<li>sex - Sex
<li>age - Age
<li>sibsp - Number of Siblings/Spouses Aboard
<li>parch - Number of Parents/Children Aboard
<li>ticket - Ticket Number
<li>fare - Passenger Fare
<li>cabin - Cabin
<li>embarked - Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)
<li>boat - Lifeboat (if survived)
<li>body - Body number (if did not survive and body was recovered)
</ul>
